# Motion Capture data

This dataset was taken from Reynolds, Zachs and Braver (2007) model of event segmentation. [[paper](https://www.ncbi.nlm.nih.gov/pubmed/21635310)] [[dataset/code]](https://wustl.app.box.com/s/0l0r2w2jg715cx567bli). The paper is worth reading butone of the main take away is that prediction errors are indicative of event boundaries, meaning that prediction errors are lower within an event than between them.

The data are 3d motion capture data of people moving. At each point in time, the position of the human in the recording is described by 18 points in (x, y, z) coordinate space. Motion is indicated by the updating of these points from timepoint to timepoint. There are 13 distinct events in the dataset and approximately 10 timepoints per event.

```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file = './datasets/motion_data.pkl'
motion_data = pd.read_pickle(data_file)
print "Number of events:", set(motion_data.EventNumber)
print "Timepoint / event:", len(motion_data) / float(len(set(motion_data.EventNumber)))
```

Formally, each scence vector $s$ is a 54D vector of the 18 points in coordinate space and each event consistest of an ordered list of scene vectors. Below is a visualization of a single event in (x, y) coordinate space. Time is represented by color, with yellow indicating early in time and red later in time (such that each of the 18 points in (x, y) space at a particular moment in time is represented with the same color)

```python
def plot_event(event_number):
    e = motion_data.loc[motion_data.EventNumber == event_number, :]
    cols = e.columns[:-1]
    e = e[cols].values

    cc = sns.color_palette("YlOrRd", n_colors=e.shape[0])
    for ii in range(e.shape[0]-1):
        for jj in range(18):
            plt.plot([e[ii, jj], e[ii+1, jj]], [e[ii, 18+jj], e[ii+1, 18+jj]],
                     color=cc[ii], marker='^', markersize=ii)
    plt.gca().set_xlabel('x-coordinate')
    plt.gca().set_ylabel('y-coordinate')
```

```python
import tensorflow as tf
from models import SEM, KerasSRN_batch
from opt.evaluate_motion import evaluate
import numpy as np

# options 
sgd_kwargs = {
    'nesterov': True, 
    'lr': 0.01, 
    'momentum': 0.9, 
#     'decay': 0.001
}

_, D = X_train.shape

# specify the model architecture (makes a big difference! especially the training parameters)
f_class = KerasSRN_batch
f_opts = dict(t=2, n_batch=250, aug_noise=0.01, sgd_kwargs=sgd_kwargs,
             n_hidden1=D, n_hidden2=D*2, hidden_act1='relu', hidden_act2='relu',)

#best: 0.64
lmda = 10 # best 100
alfa = 0.1 # best 0.01
beta = 0.09 # best 0.07
Omega = {
    'lmda': lmda,
    'alfa': alfa,
    'beta': beta, ## this parameter ends up being really important
    't': 2,
    'f_class': f_class,
    'f_opts': f_opts
}
print 'lambda =', lmda
print 'alpha =', alfa
print 'beta =', beta

r, post = evaluate(X_train, y_train, Omega, save=False)
print "Adjusted Rand Index:", r
plt.plot(post)
plt.gca().set_ylabel('Posterior cluster probability')
plt.gca().set_xlabel('Time')
plt.show()
tf.Session().close()
```

How well does the NN model were using seperate between the events? Well, our likelihood function for each event is:

$\Pr(X_{t+1}|X_{1:t}, e) = \mathcal{N}(X_t; f(X_{1:t}, \theta), \sigma I)$

One way to evaluate our NN model for events (i.e. $f(X_{1:t}, \theta)$) is to train a NN for each event (i.e. conditioned on an event) and compare the error distribution for each scene in that event vs error distribution for predicted by the NN in all of the other events. Specifically, because we can estimat the value of $\sigma$ and for an event in our sample and compare that to the estimate value of $\sigma$ for that model on all other events.

Sepearation between these lines (which may reflect overfitting) is necessary for a single event model to capture it's own dynamics while failing to capture the dynamics of the other events. 

This is highly similar to saying the within event prediction error (pe) should be lower than the between event (event boundary) pe.

```python
from scipy.stats import multivariate_normal
from scipy.optimize import minimize
from tqdm import tnrange

n_events = len(set(motion_data.EventNumber))

sigma = np.zeros(n_events)
sigma_cross = np.zeros((n_events, n_events-1))

def pull_event(event_number):
    e = motion_data.loc[motion_data.EventNumber == event_number, :]
    cols = e.columns[:-1]
    return e[cols].values


for p in tnrange(n_events):
    
    e = pull_event(p)
    
    # train a model with an event sequence, evaluate it's errors on other event sequences
    X = e[:-1]
    y = e[1:]

    N, D =  X.shape
    model = KerasSRN_batch(D, **f_opts)
    model.update(X, y)
    y_hat = model.predict(X)

    def objective(s):
        LL = 0
        for t in range(y.shape[0]):
            LL -= multivariate_normal.logpdf(y[t, :] - y_hat[t, :], 
                                             np.zeros(54), cov=s * np.eye(54))
        return LL
    
#     print y.shape, y_hat.shape
#     print objective(0.1)
#     print minimize(objective, 0.1, bounds=((1e-6, 10),)).x[0]
    sigma[p] = minimize(objective, 0.1, bounds=((1e-6, 10),)).x[0]

    events = range(n_events)
    events.remove(p)
    for t, p0 in enumerate(events):
        e0 = pull_event(p0)

        X0 = e0[:-1]
        y0 = e0[1:]
#         for t in range(shape.)
        y0_hat = model.predict(X0)
        
        def objective(s):
            LL = 0
            for t in range(y0_hat.shape[0]):
                LL -= multivariate_normal.logpdf(
                    y0[t, :] - y0_hat[t, :], np.zeros(54), cov=s * np.eye(54)
                )
            return LL
        
        sigma_cross[p, t] = minimize(objective, 0.1, bounds=((1e-6, 10),)).x[0]


with sns.axes_style('ticks'):
    fig, axes = plt.subplots(1, 2, figsize=(8, 3),
                             gridspec_kw=dict(width_ratios=[2, 1], wspace=0.4))
    
    ax = axes[0]
    ax.plot(sigma, label='correct event')

    x = range(n_events)
    y = np.mean(sigma_cross, axis=1)
    y_err = np.std(sigma_cross,  axis=1)
    ax.errorbar(x, y, y_err, label='other events')
    ax.legend()
    ax.set_xlabel('Event Number')
    ax.set_ylabel('Standard Deviation')
    sns.despine()
    
    sns.distplot(sigma, ax=axes[1], label='correct event')
    sns.distplot(sigma_cross.flatten(), ax=axes[1], label='other events')
    ax.legend()
    axes[0].set_xlabel('Standard Deviation')
```

We can see good seperation in the error distributions. We can formalize how distinguishable the models are by using the error distribution in a logistic model and plotting the ROC curve

```python
import statsmodels.api as sm

X = np.reshape(np.concatenate([sigma, sigma_cross.flatten()]), (-1, 1))
X = sm.add_constant(X)
y = np.reshape(np.concatenate([np.zeros(len(sigma)),
                    np.ones(len(sigma_cross.flatten()))]), (-1, 1))

res = sm.Logit(y, X).fit(disp=0)
y_hat = res.predict(X)

from sklearn.metrics import roc_curve, auc

fpr, tpr, thr = roc_curve(y, y_hat)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(4, 4))
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('R.O.C. curve')
plt.legend(loc="lower right")
plt.show()
```
